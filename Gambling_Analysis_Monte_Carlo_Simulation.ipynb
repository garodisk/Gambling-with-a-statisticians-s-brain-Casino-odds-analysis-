```
#Import required libraries
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#taking probability of winning as win_odds, lets take 0.486 for first few trials and then we will take input
win_odds = 0.486
```

```
#win_odds take the probability of winning, returns true if prob lies between 0 and win_odds

def win_or_lose(win_odds): 
     
    dice = random.uniform(0,1)
    
    if dice <=win_odds:
        return True

    elif dice > win_odds:
        return False
```

```
#chips_in_hand is the amount of chips a player has in hand
#betting amount is the amount a player bets in each game
#total_bets is the number of bets in one seating

def play(chips_in_hand, betting_amount, total_bets, win_odds):
    
  
    game_num = []
    updated_chips = []

    game = 1

    #lets bet for the number of bets we have in the argument and append the value of chips in hand after each bet in updated_chips list

    while game <= total_bets:
        #If we win
        if win_or_lose(win_odds):

            #Add the win amount to our chips value
            chips_in_hand = chips_in_hand + betting_amount

            game_num.append(game)
            #Append the updated_chips value after each game
            updated_chips.append(chips_in_hand)

        #If the house wins
        else:

            #Add the money to our funds
            chips_in_hand = chips_in_hand - betting_amount 
       
            game_num.append(game)
             #Append the updated_chips value after each game
            updated_chips.append(chips_in_hand)
      
        game = game + 1
    return(game_num,updated_chips,updated_chips[-1])
```

```
ending_fund= []

#lets try 5000 simulations to get a good mean value ( Monte Carlo Simulation)
#Let's start with $50000 and bet $500 bets each time for simplicity
#Lets play 5 times and see the final chips amount possible through monte carlo simulation
plt.figure(figsize=(12,6))

for x in range(1,5001):
    game_num,updated_chips,fund = play(50000 ,500, 5, win_odds)
    ending_fund.append(fund)
    plt.plot(game_num,updated_chips)

plt.title('Final chips value vs number of bets (After a player starts with $5000)')
plt.ylabel('Final chips value')
plt.xlabel('Number of bets')
plt.show()

#Lets print the statistics of the final_chips value

mean_final = sum(ending_fund)/len(ending_fund) 
variance_final_funds = sum([((x - mean_final) ** 2) for x in ending_fund]) / len(ending_fund) 
res_final = variance_final_funds ** 0.5
min_final = min(ending_fund)
max_final = max(ending_fund)

print("The player starts the game with $50000 and ends with $" + str(mean_final))
print("The standard deviation is $" + str(res_final))
print('The minimum anad maximum funds that a player can possible end with are: {} and {}'.format(min_final,max_final))
```

```
plt.figure(figsize=(12,6))
sns.distplot(ending_fund)
plt.title('Distribution of final chips value in each simulation')
plt.ylabel('Probability')
plt.xlabel('Final chips value')
plt.show()
```

```
#this function will simulate bets and return the mean, standard deviation, minim and maxixmum
#input - initial money, each bet, games_played, win_odds, num_simulations


def simulate_bets(initial_money, each_bet, games_played, win_odds, num_simulations):
  
  ending_fund= []

  for x in range(1,num_simulations+1):
    game_num,updated_chips,fund = play(initial_money, each_bet, games_played, win_odds)
    ending_fund.append(fund)
  
  mean_final = sum(ending_fund)/len(ending_fund) 
  variance_final_funds = sum([((x - mean_final) ** 2) for x in ending_fund]) / len(ending_fund) 
  std_final = variance_final_funds ** 0.5
  min_final = min(ending_fund)
  max_final = max(ending_fund)
  return(mean_final, std_final, min_final, max_final)
```

```
win_odds_list = np.linspace(.45, .495, num=100)
```

```
win_odds_list
```

```
col_names =  ['win_odds', 'mean_final', 'std_final', 'min_final', 'max_final']

final_coins  = pd.DataFrame(columns = col_names)

for odds in win_odds_list:
  
  mean_val, std_val, min_val, max_val = simulate_bets(50000, 500, 1000, odds, 5000)
  final_coins = final_coins.append({'win_odds': odds, 'mean_final': mean_val, 'std_final': std_val, 'min_final':min_val , 'max_final':max_val }, ignore_index=True)
```
